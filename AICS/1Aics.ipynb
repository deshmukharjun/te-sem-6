```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("C:/Users/pushk/downloads/spam.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data['Category'] = data['Category'].map({ 'ham' : 1 , 'spam'  : 0})
data.head()
```

```python
df = data
```

```python
lb = LabelEncoder()
df['Category'] = lb.fit_transform(df['Category'])
```

```python
df.head()
```

```python
X = data['Message']   # X = data.drop(['xyz'], axis=1)
Y = data['Category']   # Y = data['xyz']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X , Y ,test_size = 0.3, random_state = 23)
```

```python
vec = CountVectorizer()                       #scaler = StandardScaler()
X_train_vec = vec.fit_transform(X_train)     # X_scaler = scaler.fit_transform(X)
X_test_vec = vec.transform(X_test)
```

```python
nb = MultinomialNB()                          # model = RandomForestClassifier(n_estimators = 100, random_state=69)
nb.fit(X_train_vec, Y_train)                     #model.fit(X_train, Y_train)
```

```python
pred = nb.predict(X_test_vec)                     #pred = model.predict(X_test)
print("accuracy Score", accuracy_score(Y_test,pred))
```
